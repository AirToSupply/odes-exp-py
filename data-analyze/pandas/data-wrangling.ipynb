### 2.数据连接

```python
left = pd.DataFrame({
    "A": ["A0", "A1", "A2", "A3"],
    "B": ["B0", "B1", "B2", "B3"],
    "key": ["K0", "K1", "K2", "K3"]
})
left
```

```python
right = pd.DataFrame({
    "C": ["C0", "C1", "C2", "C3"],
    "D": ["D0", "D1", "D2", "D3"],
    "key": ["K0", "K1", "K2", "K3"]
})
right
```

#### merge的使用

```python
# 默认情况下连接的数据集有相同的key则会自动连接
pd.merge(left, right)
```

```python
# 通过on参数指定连接键
pd.merge(left, right, on = 'key')
```

```python
l2 = pd.DataFrame({
    "A": ["A0", "A1", "A2", "A3"],
    "B": ["B0", "B1", "B2", "B3"],
    "k1": ["K0", "K0", "K2", "K3"],
    "k2": ["K0", "K1", "K2", "K3"]
})
l2
```

```python
r2 = pd.DataFrame({
    "C": ["C0", "C1", "C2", "C3"],
    "D": ["D0", "D1", "D2", "D3"],
    "k1": ["K0", "K1", "K2", "K3"],
    "k2": ["K0", "K1", "K2", "K3"]
})
r2
```

```python
# 指定多个建连接
# 没有匹配成功的则删除不进行连接
pd.merge(l2, r2, on = ["k1", "k2"])
```

```python
l3 = pd.DataFrame({
    "A": ["A0", "A1", "A2", "A3"],
    "B": ["B0", "B1", "B2", "B3"],
    "k1": ["K0", "K0", "K1", "K2"],
    "k2": ["K0", "K1", "K0", "K1"]
})
l3
```

```python
r3 = pd.DataFrame({
    "C": ["C0", "C1", "C2", "C3"],
    "D": ["D0", "D1", "D2", "D3"],
    "k1": ["K0", "K1", "K1", "K2"],
    "k2": ["K0", "K0", "K0", "K0"]
})
r3
```

```python
# 通过参数how指定连接方式
# 左链接
pd.merge(l3, r3, how = "left", on = ["k1", "k2"])
```

```python
# 通过参数how指定连接方式
# 右链接
pd.merge(l3, r3, how = "right", on = ["k1", "k2"])
```

```python
# 通过参数how指定连接方式
# 默认不指定链接方式则是内连接
pd.merge(l3, r3, how = "inner", on = ["k1", "k2"])
```

```python
# 通过参数how指定连接方式
# 全外连接
pd.merge(l3, r3, how = "outer", on = ["k1", "k2"])
```

```python
l4 = pd.DataFrame({
    "data": [1, 7, 1, 2, 8, 5, 0],
    "key": ['b', 'b', 'a', 'c', 'a', 'a', 'b']
})
l4
```

```python
r4 = pd.DataFrame({
    "data": [0, 4, 7],
    "key": ['a', 'b', 'd']
})
r4
```

```python
# 如何处理重复列
# 默认为：_x和_y
pd.merge(l4, r4, on = "key")
```

```python
# 通过suffixes参数指定重复列的名称？
pd.merge(l4, r4, on = "key", suffixes = ["_left", "_right"])
```

```python
l5 = pd.DataFrame({
    "data": [1, 7, 1, 2, 8, 5, 0],
    "key": ['b', 'b', 'a', 'c', 'a', 'a', 'b']
})
l5
```

```python
r5 = pd.DataFrame({
    "data": [0, 4, 7]
}, index = ['a', 'b', 'd'])
r5
```

```python
# 将左集的key和有集的索引列进行连接
# 参数left_on指定左集连接的键是哪个列
# 参数right_index指定右集连接的键是当前数据集的行索引
pd.merge(l5, r5, left_on = "key", right_index = True)
```

```python
l6 = pd.DataFrame({
    "Chinese": [1, 3, 5],
    "Math": [2, 4, 6]
}, index = list("ace"))
l6
```

```python
r6 = pd.DataFrame({
    "Chinese": [7, 9, 11, 13],
    "Math": [8, 10, 12, 14]
}, index = list("bcde"))
r6
```

```python
# 通过数据集的行索引进行连接
pd.merge(l6, r6, how = "outer", left_index = True, right_index = True)
```

#### concat

```python
df1 = pd.DataFrame(np.arange(6).reshape(3, 2), index = list("abc"), columns = ["one", "two"])
df1
```

```python
df2 = pd.DataFrame(np.arange(4).reshape(2, 2) + 5, index = list("ac"), columns = ["three", "four"])
df2
```

```python
# 默认按照横轴方向合并
# 默认外连接
pd.concat([df1, df2])
```

```python
# 按照竖轴方向合并
pd.concat([df1, df2], axis = 1)
```

### 数据重塑和轴线旋转

#### 重塑层次化索引

##### 1.stack函数可以将列索引转变成行索引，完成层级索引，即可以将DataFrame转换为Series对象

```python
df3 = pd.DataFrame(
    np.random.randint(80, 100, 6).reshape(2, 3), 
    index = pd.Index(list("AB"), name = "Name"),
    columns = pd.Index(["Chinese", "Math", "English"], name = "Course")
)
df3
```

```python
print("<df3 type>", type(df3))
series = df3.stack()
print("<series type>", type(series))
series
```

##### 2.unstack函数是stack函数逆操作，可以将层级索引展开（默认操作内层索引，即：level=-1），可以将Series转换为DataFrame对象

```python
_df = series.unstack()
print("<_df type>", type(_df))
_df
```

```python
series.unstack(level = 0)
```

#### 轴向旋转

```python
data = pd.DataFrame({
     'Date': ['08-24', '08-24', '08-25', '08-25', '08-26', '08-26'],
     'No': ['A', 'B', 'C', 'A', 'B', 'C'],
     'HJ89': [5, 2, 3, 1, 5, 6],
     'VT56': [4, 2, 3, 1, 4, 1]})
data
```

```python
# 将列Date作为数据透视的行索引
# 将列No作为数据透视的列索引
# 将列HJ89作为数据透视的对象
data.pivot(index='Date', columns='No', values='HJ89')
```

```python
# 将列Date作为数据透视的行索引
# 将列No作为数据透视的列索引
# 将列VT56作为数据透视的对象
data.pivot(index='Date', columns='No', values='VT56')
```
