### 2.数据连接

```python
left = pd.DataFrame({
    "A": ["A0", "A1", "A2", "A3"],
    "B": ["B0", "B1", "B2", "B3"],
    "key": ["K0", "K1", "K2", "K3"]
})
left
```

```python
right = pd.DataFrame({
    "C": ["C0", "C1", "C2", "C3"],
    "D": ["D0", "D1", "D2", "D3"],
    "key": ["K0", "K1", "K2", "K3"]
})
right
```

#### merge的使用

```python
# 默认情况下连接的数据集有相同的key则会自动连接
pd.merge(left, right)
```

```python
# 通过on参数指定连接键
pd.merge(left, right, on = 'key')
```

```python
l2 = pd.DataFrame({
    "A": ["A0", "A1", "A2", "A3"],
    "B": ["B0", "B1", "B2", "B3"],
    "k1": ["K0", "K0", "K2", "K3"],
    "k2": ["K0", "K1", "K2", "K3"]
})
l2
```

```python
r2 = pd.DataFrame({
    "C": ["C0", "C1", "C2", "C3"],
    "D": ["D0", "D1", "D2", "D3"],
    "k1": ["K0", "K1", "K2", "K3"],
    "k2": ["K0", "K1", "K2", "K3"]
})
r2
```

```python
# 指定多个建连接
# 没有匹配成功的则删除不进行连接
pd.merge(l2, r2, on = ["k1", "k2"])
```

```python
l3 = pd.DataFrame({
    "A": ["A0", "A1", "A2", "A3"],
    "B": ["B0", "B1", "B2", "B3"],
    "k1": ["K0", "K0", "K1", "K2"],
    "k2": ["K0", "K1", "K0", "K1"]
})
l3
```

```python
r3 = pd.DataFrame({
    "C": ["C0", "C1", "C2", "C3"],
    "D": ["D0", "D1", "D2", "D3"],
    "k1": ["K0", "K1", "K1", "K2"],
    "k2": ["K0", "K0", "K0", "K0"]
})
r3
```

```python
# 通过参数how指定连接方式
# 左链接
pd.merge(l3, r3, how = "left", on = ["k1", "k2"])
```

```python
# 通过参数how指定连接方式
# 右链接
pd.merge(l3, r3, how = "right", on = ["k1", "k2"])
```

```python
# 通过参数how指定连接方式
# 默认不指定链接方式则是内连接
pd.merge(l3, r3, how = "inner", on = ["k1", "k2"])
```

```python
# 通过参数how指定连接方式
# 全外连接
pd.merge(l3, r3, how = "outer", on = ["k1", "k2"])
```

```python
l4 = pd.DataFrame({
    "data": [1, 7, 1, 2, 8, 5, 0],
    "key": ['b', 'b', 'a', 'c', 'a', 'a', 'b']
})
l4
```

```python
r4 = pd.DataFrame({
    "data": [0, 4, 7],
    "key": ['a', 'b', 'd']
})
r4
```

```python
# 如何处理重复列
# 默认为：_x和_y
pd.merge(l4, r4, on = "key")
```

```python
# 通过suffixes参数指定重复列的名称？
pd.merge(l4, r4, on = "key", suffixes = ["_left", "_right"])
```

```python
l5 = pd.DataFrame({
    "data": [1, 7, 1, 2, 8, 5, 0],
    "key": ['b', 'b', 'a', 'c', 'a', 'a', 'b']
})
l5
```

```python
r5 = pd.DataFrame({
    "data": [0, 4, 7]
}, index = ['a', 'b', 'd'])
r5
```

```python
# 将左集的key和有集的索引列进行连接
# 参数left_on指定左集连接的键是哪个列
# 参数right_index指定右集连接的键是当前数据集的行索引
pd.merge(l5, r5, left_on = "key", right_index = True)
```
